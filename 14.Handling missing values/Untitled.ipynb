```python
import pandas as pd
ufo = pd.read_table('ufo.csv', sep=',')
ufo.head()
```

```python
ufo.isnull().tail()
```

```python
ufo.isnull().sum()
#true is counted as 1 and false as 0. Therefore sum gives a value
```

```python
ufo[ufo.City.isnull()]
```

```python
#drop the row if any of its values are missing
ufo.dropna(how='any')
```

```python
ufo.count()
```

```python
ufo.dropna(how='any').count()
```

```python
#drop the row if all of its values are missing
ufo.dropna(how='all')
```

```python
ufo.dropna(how='all').count()
```

```python
#drop the row if any of specified values are missing
ufo.dropna(subset=['City', 'Colors Reported'], how='any')
```

```python
ufo.dropna(subset=['City', 'Colors Reported'], how='any').count()
```
