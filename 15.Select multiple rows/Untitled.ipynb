```python
import pandas as pd
ufo = pd.read_table('ufo.csv', sep=',')
ufo.head()
```

```python
# selecting rows and filtering columns by labels
ufo.loc[0, :]
#Row zero and all columns
```

```python
ufo.loc[[0,1,2,3,4], :]
#Rows 0,1,2,3,4 and all columns
```

```python
ufo.loc[0:4, :]
#Rows 0,1,2,3,4 and all columns
```

```python
ufo.loc[:,['City','State']]
```

```python
ufo.loc[ufo.City=='Oakland',:]
```

```python
#filtering rows and selecting columns by integer position
ufo.iloc[[0,5],:]
```

```python
ufo.iloc[0:10,:]
```
