```python
import pandas as pd
titanic = pd.read_table('titanic.csv',sep=',', index_col='PassengerId')
titanic.head()
```

```python
feature=['Pclass','Parch']
```

```python
x=titanic.loc[:,feature]
```

```python
#target vector
y=titanic.Survived
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x,y)
```

```python
test = pd.read_table('titanic_test.csv', sep=',')
test.head()
```

```python
xx = test.loc[:,feature]
```

```python
xx.shape
```

```python
new_predict_class = lr.predict(xx)
```

```python
pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': new_predict_class}).set_index('PassengerId')
```

```python
pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': new_predict_class}).set_index('PassengerId').to_csv('output.csv')
```
