```python
import pandas as pd
drinks = pd.read_table('drinks.csv', sep=',')
drinks.head()
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.min()
```

```python
drinks.groupby('continent').beer_servings.max()
```

```python
drinks.groupby('continent').beer_servings.agg(['min', 'max', 'count', 'mean'])
```

```python
%matplotlib inline
drinks.groupby('continent').beer_servings.agg(['min', 'max', 'count', 'mean']).plot(kind='bar')
drinks.groupby('continent').beer_servings.agg(['min', 'max', 'count', 'mean']).plot(kind='hist')
```
