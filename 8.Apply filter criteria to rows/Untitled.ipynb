# Apply multiple filter criteria to Data Frame

```python
import pandas as pd
imd = pd.read_table('imdb.csv', sep=',')
imd.head()
```

```python
imd[imd.duration>=210]
```

```python
imd[(imd.duration>=210) & (imd.genre == 'Crime')]
```

```python
# Or operator
imd[imd.genre.isin(['Crime','Adventure'])]
```
